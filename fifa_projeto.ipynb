# Valor de mercado dos jogadores da LaLiga, somados por Clube

```python
fifa['Value'].describe()
```

```python
fifa['Value'] = fifa['Value'].str.replace('[/€\,]', '', regex=True)
```

```python
def convert_value(value):
    if 'K' in value:
        return float(value.replace('K', '')) / 1000
    elif 'M' in value:
        return float(value.replace('M', ''))
    else:
        return float(value)
```

```python
fifa['Value'] = fifa['Value'].apply(convert_value)
```

```python
fifa['Value'].describe()
```

```python
fifa.head(50)
```

```python
fifa.tail()
```

```python
fifa_laliga = fifa[fifa['League'] == 'LALIGA SANTANDER']
```

```python
fifa_laliga_total = fifa_laliga.groupby('Club')['Value'].sum().reset_index().sort_values(by = 'Value', ascending = False)
```

```python
fifa_laliga_total
```

```python
plt.figure(figsize = (16, 6))

sns.barplot(data = fifa_laliga_total.sort_values(by = 'Value', ascending = False), 
            x = 'Club', 
            y = 'Value',
            palette="flare")

plt.title('Soma do Valor de Mercado dos jogadores da LaLiga, por Clubes')
plt.xticks(rotation = 80)
plt.show()
```

# Quais os jogadores com as maiores Clausulas de Contrato do Man City

```python
fifa['Release Clause'] = fifa['Release Clause'].str.replace('[/€\,]', '', regex=True)
```

```python
def convert_clause(value):
    if 'K' in value:
        return float(value.replace('K', '')) / 1000
    elif 'M' in value:
        return float(value.replace('M', ''))
    elif 'No Release Clause' in value:
        return float(value.replace('No Release Clause', '0'))
    else:
        return float(value)
```

```python
fifa['Release Clause'] = fifa['Release Clause'].apply(convert_clause)
```

```python
fifa_club = fifa[fifa['Club'] == 'Manchester City']
```

```python
fifa_clause = fifa_club.groupby('Name')['Release Clause'].sum().reset_index().sort_values(by = 'Release Clause', ascending = False)
```

```python
plt.figure(figsize = (16, 6))

sns.barplot(data = fifa_clause, 
            x = 'Name', 
            y = 'Release Clause',
           palette='blend:lightblue,navy')

plt.title('Valor da Clausula de Contrato dos jogadores do Manchester City')
plt.grid(axis='y', linestyle='--', color='gray', linewidth=0.5)
plt.xticks(rotation = 80)
plt.show()
```

# Em quais clubes estão os 20 melhores jogadores brasileiros do FIFA19

```python
fifa_brazil = fifa[fifa['Nationality'] == 'Brazil']
```

```python
fifa_brazil_top = fifa_brazil.groupby(['Club','Name']).max().reset_index().sort_values(by = 'Overall', ascending = False).head(25)
```

```python
fifa_brazil_top = fifa_brazil_top.sort_values(by = 'Club').reset_index()
```

```python
fifa_brazil_top
```

```python
brazil = fifa_brazil_top.groupby(['Club'])['Name'].count().reset_index()
brazil
```
